# Mean
The mean is the average value of a dataset. It is calculated by summing up all the values in the dataset and dividing by the total number of values.
The mean is widely used in various fields, such as statistics, mathematics, and science.
It provides a measure of central tendency and is useful for understanding the typical or average value in a dataset.
The mean is sensitive to extreme values, known as outliers, which can significantly influence its value.


```python
# Generate a large dataset with random values
np.random.seed(42)  # for reproducibility
dataset_size = 1000
data = np.random.normal(loc=50, scale=10, size=dataset_size)

# Calculate the mean using numpy
mean = np.mean(data)

# Create a histogram (distribution graph) of the data
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

# Add a vertical line for the mean
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean:.2f}')

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution with Mean Line")
plt.legend()

# Show the graph
plt.show()
```

# Median

The median represents the middle value of a dataset when it is sorted in ascending or descending order.
To find the median, the dataset is arranged in order, and the middle value is selected. If the dataset has an odd number of values, the middle value is the median. If the dataset has an even number of values, the median is the average of the two middle values.
The median is a robust measure of central tendency that is not affected by outliers.
It is commonly used when the dataset contains extreme values or when the distribution is skewed.

```python
# Generate a large dataset with random values
np.random.seed(42)  # for reproducibility
dataset_size = 1000
data = np.random.normal(loc=50, scale=10, size=dataset_size)

# Calculate the median using numpy
median = np.median(data)

# Create a histogram (distribution graph) of the data
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

# Add a vertical line for the median
plt.axvline(median, color='orange', linestyle='dashed', linewidth=2, label=f'Median = {median:.2f}')

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution with Median Line")
plt.legend()

# Show the graph
plt.show()
```

# Mean VS Medain in right-skewed data

```python
np.random.seed(42)  # for reproducibility
dataset_size = 1000

# Generate right-skewed data using gamma distribution
shape = 2  # Shape parameter for gamma distribution (higher values increase skewness)
scale = 10  # Scale parameter for gamma distribution
data = np.random.gamma(shape, scale, dataset_size)

# Calculate the mean and median
mean = np.mean(data)
median = np.median(data)

# Plot the histogram to visualize the data distribution
plt.hist(data, bins='auto', alpha=0.7, rwidth=0.85)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median, color='g', linestyle='dashed', linewidth=1, label='Median')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Right-Skewed Distribution')
plt.legend()

# Display the plot
plt.show()
print('The mean are shift to the right because the outliers')
print("Mean:", mean)
print("Median:", median)
```

# Mode
The mode is the value that appears most frequently in a dataset.
The mode is determined by counting the occurrences of each value in the dataset and identifying the value with the highest frequency.
It is often used to identify the most common or popular value in a dataset.
The mode can be useful for categorical or discrete data, such as survey responses or types of objects.

```python
from collections import Counter

# Example discrete dataset 
dataset = [5, 2, 7, 10, 5, 3, 8, 5]

# Calculate the mode
counts = Counter(dataset)
mode = max(counts, key=counts.get)

# Create a bar graph to visualize the mode
values = counts.keys()
frequencies = counts.values()

plt.bar(values, frequencies, color='blue')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Mode of Dataset')

# Display the plot
plt.show()
```
